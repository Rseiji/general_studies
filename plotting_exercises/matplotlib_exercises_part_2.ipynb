# Exercise 1: Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: â€“

X label name = Month Number
Y label name = Total profit

````python
df.plot(x='month_number', y='total_profit')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company profit per month')
````

# Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line-color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


````python
df.plot(x='month_number', y='total_profit', ls='--', color='red', lw=3, marker='o', markerfacecolor='black')
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company profit per month')
````

# Exercise 3: Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
df.plot(
    x='month_number',
    y=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'],
    color=['blue', 'orange', 'green', 'red', 'purple', 'brown'],
    lw=3,
    marker='o',
    figsize=(10,5)
)
plt.legend(loc='upper left')
plt.xlabel('Month Number', fontsize=15)
plt.ylabel('Sales units in number', fontsize=15)
plt.title('Sales Data', fontsize=25)
````

# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot. gridline style should â€œâ€“â€œ.

The scatter plot should look like this.

````python
plt.scatter(df['month_number'].tolist(), df['toothpaste'].tolist())
plt.xlabel('Month Number', fontsize=15)
plt.ylabel('Nmuber of units sold', fontsize=15)
plt.title('Tooth Paste Sales Data', fontsize=20)
plt.legend()
````

# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.



````python
df
````

````python
sns.set_style('darkgrid')

fig, axs = plt.subplots(figsize=(10,7))

axs.bar(df['month_number']+0.1, df['facecream'], width=0.2, color='blue', label='Face Cream sales data')
axs.bar(df['month_number']-0.1, df['facewash'], width=0.2, color='yellow', label='Face Wash sales data')
axs.set_title('Facewash and facecream sales data', fontsize=14)
axs.tick_params(axis='both', labelsize=15)
axs.legend(loc='upper left', fontsize=12)
axs.set_xlabel('Month Number', fontsize=15)
axs.set_ylabel('Sales Units in Number', fontsize=15)
````

# Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges


````python
fig, axs = plt.subplots(figsize=(10,5))
axs.hist(df['total_profit'])
axs.legend('Profit Data')
axs.set_xlabel('profit range in dollar')
axs.set_ylabel('Actual Profit in dollar')
````

# Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

Note: In Pie chart, display number of units sold per year of each product in percentage.

````python
categories_sum = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
fig, axs = plt.subplots(figsize=(7,7))
axs.pie(categories_sum, autopct='%1.1f%%', labels=categories_sum.index)
axs.set_title('Sales Data', fontsize=15)
axs.legend(loc='lower right')
````

# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

````python
fig, axs = plt.subplots(2,1, figsize=(20,7), sharex=True)
axs[0].plot(df['month_number'], df['bathingsoap'], color='black', marker='o', lw=3)
axs[1].plot(df['month_number'], df['facewash'], color='limegreen', marker='o', lw=3)

axs[0].set_title('Sales data of a Bathingsoap', fontsize=15)
axs[1].set_title('Sales data of a facewash', fontsize=15)

axs[1].set_xlabel('Month NUmber', fontsize=15)

axs[1].set_ylabel('Sales units in number', fontsize=15)
````

# Exercise Question 10: Read all product sales data and show it using the stack plot

````python
fig, axs = plt.subplots(figsize=(20,7))
axs.stackplot(
    df['month_number'], df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer'],
    labels=['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
)
axs.legend(loc='upper left')
axs.set_xlabel('Month Number', fontsize=15)
axs.set_ylabel('Sales Units in Number', fontsize=15)
axs.set_title('All product sales data using stackplot', fontsize=20)
````
