# Heatmap

```python
fig, axs = plt.subplots(figsize=(15,15))
sns.heatmap(df.corr(), ax=axs, annot=True, fmt='.1f', linewidths=5, linecolor='blue')
axs.tick_params(labelsize=15)
```

# Histogram

```python
sns.set_style('darkgrid')
sns.set_palette("RdPu")
f = sns.histplot(x=df['Age'], bins=40)
ax = f.axes
ax.set_xlabel('Age Value', fontsize=14)
ax.set_ylabel('Age Qtd Count', fontsize=14)
ax.set_title('My Histogram', fontsize=14)
```

# Barplot

```python
fig, ax = plt.subplots(figsize=(7,5))
sns.barplot(x=df['EducationField'], y=df['DistanceFromHome'], ax=ax, ci=None)
ax.tick_params(rotation=90)
```

```python
sns.barplot(y=df['EducationField'], x=df['DistanceFromHome'], ci=None, label='distance_home', color='blue', alpha=0.5)
sns.barplot(y=df['EducationField'], x=df['Education'], ci=None, label='education', color='red', alpha=0.3)
sns.barplot(y=df['EducationField'], x=df['WorkLifeBalance'], ci=None, label='education', color='green', alpha=1)
```

# Boxplot

```python
fig, ax = plt.subplots(figsize=(10,7))
sns.boxplot(x=df['EducationField'], y=df['YearsSinceLastPromotion'], ax=ax, hue=df['MaritalStatus'], palette='RdPu')
ax.tick_params(rotation=45)
```

# Pairplot

```python
sns.pairplot(df[['Age', 'BusinessTravel', 'DailyRate',  'Department', 'RelationshipSatisfaction', 'TotalWorkingYears']], hue='BusinessTravel', height=3)
```

# PairGrid

```python
# Basic
x_vars = ['MonthlyIncome', 'EnvironmentSatisfaction', 'PercentSalaryHike', 'RelationshipSatisfaction']#, 'StockOptionLevel', 'TotalWorkingYears', 'WorkLifeBalance']
y_vars = ['MonthlyIncome', 'EnvironmentSatisfaction', 'PercentSalaryHike', 'RelationshipSatisfaction']#, 'StockOptionLevel', 'TotalWorkingYears', 'WorkLifeBalance']
g = sns.PairGrid(df, x_vars=x_vars, y_vars=y_vars, height=2.5)
g.map(sns.scatterplot)
```

```python
# Row
x_vars = ['MonthlyIncome', 'EnvironmentSatisfaction', 'PercentSalaryHike', 'RelationshipSatisfaction']
y_vars = ['MonthlyIncome']
g = sns.PairGrid(df, x_vars=x_vars, y_vars=y_vars)
g.map(sns.scatterplot)
```

```python
# Col
x_vars = ['MonthlyIncome']
y_vars = ['MonthlyIncome', 'EnvironmentSatisfaction', 'PercentSalaryHike', 'RelationshipSatisfaction']
g = sns.PairGrid(df, x_vars=x_vars, y_vars=y_vars)
g.map(sns.scatterplot)
```

```python
x_vars = ['MonthlyIncome', 'EnvironmentSatisfaction', 'PercentSalaryHike', 'RelationshipSatisfaction']#, 'StockOptionLevel', 'TotalWorkingYears', 'WorkLifeBalance']
y_vars = ['MonthlyIncome', 'EnvironmentSatisfaction', 'PercentSalaryHike', 'RelationshipSatisfaction']#, 'StockOptionLevel', 'TotalWorkingYears', 'WorkLifeBalance']
g = sns.PairGrid(df, x_vars=x_vars, y_vars=y_vars, hue='EducationField')
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.add_legend()
```

# FacetGrid

```python
g = sns.FacetGrid(df, col='Gender', row='Education', hue='EducationField')
g.map(sns.scatterplot, 'BusinessTravel', 'YearsAtCompany')
g.add_legend()
```
