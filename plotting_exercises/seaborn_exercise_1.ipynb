# Heatmap

```python
fig, axs = plt.subplots(figsize=(15,15))
sns.heatmap(df.corr(), ax=axs, annot=True, fmt='.1f', linewidths=5, linecolor='blue')
axs.tick_params(labelsize=15)
```

# Histogram

```python
sns.set_style('darkgrid')
sns.set_palette("RdPu")
f = sns.histplot(x=df['Age'], bins=40)
ax = f.axes
ax.set_xlabel('Age Value', fontsize=14)
ax.set_ylabel('Age Qtd Count', fontsize=14)
ax.set_title('My Histogram', fontsize=14)
```

# Barplot

```python
fig, ax = plt.subplots(figsize=(7,5))
sns.barplot(x=df['EducationField'], y=df['DistanceFromHome'], ax=ax, ci=None)
ax.tick_params(rotation=90)
```

```python
sns.barplot(y=df['EducationField'], x=df['DistanceFromHome'], ci=None, label='distance_home', color='blue', alpha=0.5)
sns.barplot(y=df['EducationField'], x=df['Education'], ci=None, label='education', color='red', alpha=0.3)
sns.barplot(y=df['EducationField'], x=df['WorkLifeBalance'], ci=None, label='education', color='green', alpha=1)
```
